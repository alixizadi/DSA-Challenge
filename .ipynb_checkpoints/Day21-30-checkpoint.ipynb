## 21. 3Sum `Medium`
- Given an integer array nums, return all the triplets `[nums[i], nums[j], nums[k]]` such that `i != j , i != k and j != k` , and nums[i] + nums[j] + nums[k] == 0
- Notice that the solution set must not contain duplicate triplets.


````python
def get_3sum(nums, s):
    if len(nums) < 0:
        return []
    l = len(nums)
    output = []
    for i in range(0,l):
        for j in range(i + 1, l):
            for k in range(j + 1, l):
                if nums[i] + nums[j] + nums[k] == s:
                    output.append((nums[i],nums[j],nums[k]))
    return output
print(get_3sum([1,2,3,4,5,6,7,8,9], 19))
````

## 22. Binary Tree Level Order Traversal `Medium`
- Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

````python
class TreeNode():
    def __init__(self, data,left = None, right = None):
        self.data = data
        self.left = left
        self.right = right
        
    def addChild(self, c):
        if c.data < self.data:
            if self.left == None:
                self.left = c
            else:
                self.left.addChild(c)                
        elif c.data > self.data:
            if self.right == None:
                self.right = c
            else:
                self.right.addChild(c)                
        else:
            raise ValueError("Duplicate value")
    
    def __str__(self):
        print(self.data)
        if self.left != None:
            print(self.left)
        if self.right != None:
            print(self.right)
        return '' 
    
def level_traverse(nodes):
    if len(nodes) == 0:
        return
    
    queue = []
    for node in nodes:
        yield node.data
        if node.left is not None:
            queue.append(node.left)
        if n.right is not None:
            queue.append(node.right)
            
    yield from level_traverse(queue)


root = TreeNode(10)
root.addChild(TreeNode(5))
root.addChild(TreeNode(12))
root.addChild(TreeNode(7))
root.addChild(TreeNode(2))
root.addChild(TreeNode(15))
root.addChild(TreeNode(22))


"""
       10
      / \
     5   12
    / \    \
   2   7    15
              \
               22
"""

for n in  level_traverse([root]) :
    print(n)
````

## 23. Clone Graph `Medium`
- Given a reference of a node in a connected undirected graph, Return a deep copy (clone) of the graph.


````python
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def cloneGraph(node):
    if not node:
        return None

    visited = {}
    return clone_node(node, visited)

def clone_node(node, visited):
    if node in visited:
        return visited[node]

    cloned_node = Node(node.val)
    visited[node] = cloned_node

    for neighbor in node.neighbors:
        cloned_neighbor = clone_node(neighbor, visited)
        cloned_node.neighbors.append(cloned_neighbor)

    return cloned_node
````

````python
# Create the original graph
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.neighbors = [node2, node4]
node2.neighbors = [node1, node3]
node3.neighbors = [node2, node4]
node4.neighbors = [node1, node3]

cloned_graph = cloneGraph(node1)

def print_graph(node):
    if not node:
        return
    
    visited = set()
    queue = [node]
    
    while queue:
        curr_node = queue.pop(0)
        
        if curr_node in visited:
            continue
        
        visited.add(curr_node)
        print(f"Node: {curr_node.val}")
        print("Neighbors: ", end="")
        for neighbor in curr_node.neighbors:
            print(f"{neighbor.val} ", end="")
            queue.append(neighbor)
        print()

print_graph(cloned_graph)
````

## 24. Evaluate Reverse Polish Notation `Medium`
- You are given an array of strings tokens that represents an arithmetic expression in a Reverse Polish Notation.
- Evaluate the expression. Return an integer that represents the value of the expression.

> Here's an example to illustrate the difference between infix notation (conventional notation) and Reverse Polish Notation:
>
> Infix Notation: 3 + 4 * 2
>
> Reverse Polish Notation (RPN): 3 4 2 * +
>


````python
def evaluate(expression):
    stack = []

    operators = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '*': lambda x, y: x * y,
        '/': lambda x, y: x / y
    }

    for token in expression:
        if token in operators:
            # Token is an operator
            operand2 = stack.pop()
            operand1 = stack.pop()
            result = operators[token](operand1, operand2)
            stack.append(result)
        else:
            # Token is an operand
            stack.append(float(token))

    return stack[0]
````

````python
expression = ['3', '4', '2', '*', '+']
result = evaluate(expression)
print(result)  # Output: 11.0
expression = ['3', '4', '2','7','/', '-', '+']
result = evaluate(expression)
print(round(result,2))  # Output: 6.71
````

## 25. Course Schedule `Medium`
- There are a total of numCourses courses you have to take, labeled from 0 to numCourses 1 .
- You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai .
- Return true if you can finish all courses. Otherwise, return false .

```python
numCourses = 2
prerequisites = [[1, 0]]
# Expected output: True 
```

```python
numCourses = 2
prerequisites = [[1, 0], [0, 1]]
# Expected output: False
```

Test Case 3:
```python
numCourses = 4
prerequisites = [[1, 0], [2, 1], [3, 2]]
# Expected output: True
```

Test Case 4:
```python
numCourses = 3
prerequisites = [[1, 0], [2, 0]]
# Expected output: True
```


````python
class Course:
    def __init__(self, data):
        self.data = data
        self.neighbors = []

    def add_neighbor(self, ng):
        self.neighbors.append(ng)

    def __str__(self):
        return str(self.data)

    def __repr__(self):
        return f"Course({self.data}, {self.neighbors})"

    def __eq__(self, other):
        if isinstance(other, Course):
            return self.data == other.data
        return False

    def __hash__(self):
        return hash(self.data)


class CourseGraph:
    def __init__(self, prerequisites):
        self.data = {}
        for p in prerequisites:
            course_a = self.data.setdefault(p[1], Course(p[1]))
            course_b = self.data.setdefault(p[0], Course(p[0]))
            course_a.add_neighbor(course_b)

    def has_cycle(self):
        if len(self.data) == 0:
            return False

        visited = set()
        for start in self.data.values():
            if start.data not in visited:
                if self.traverse(start, visited):
                    return True

        return False

    def traverse(self, start, visited):
        visited.add(start.data)
        for n in start.neighbors:
            if n.data in visited or self.traverse(n, visited):
                return True

        visited.remove(start.data)
        return False

    def __str__(self):
        return str(self.data)
````

````python
prerequisites = [[1, 0], [2, 1], [3, 2], [3, 1]]
g = CourseGraph(prerequisites)
# Expected output: True 
print(not g.has_cycle())


prerequisites = [[1, 0]]
g = CourseGraph(prerequisites)
# Expected output: True 
print(not g.has_cycle())

prerequisites = [[1, 0], [0, 1]]
g = CourseGraph(prerequisites)
# Expected output: False 
print(not g.has_cycle())

prerequisites = [[1, 0], [2, 1], [3, 2]]
g = CourseGraph(prerequisites)
# Expected output: True 
print(not g.has_cycle())

prerequisites = [[1, 0], [2, 1], [0, 2]]
g = CourseGraph(prerequisites)
# Expected output: False 
print(not g.has_cycle())
````

## 26. Implement Trie (Prefix Tree) `Medium`

- Implement the Trie class:

- Trie() Initializes the trie object.
- void insert(String word) Inserts the string word into the trie.
- boolean search(String word) Returns true if the string word is in the trie boolean startsWith(String prefix) Returns true if there is a previously
- inserted string word that has the prefix.


````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_word = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_word = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_word

    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True

    def delete(self, word):
        def _delete_helper(node, word, index):
            if index == len(word):
                if node.is_word:
                    node.is_word = False
                    return len(node.children) == 0
                return False
            char = word[index]
            if char not in node.children:
                return False
            child_node = node.children[char]
            delete_child = _delete_helper(child_node, word, index + 1)
            if delete_child:
                del node.children[char]
            return len(node.children) == 0 and not node.is_word and delete_child

        _delete_helper(self.root, word, 0)
````

````python
trie = Trie()
trie.insert("apple")
trie.insert("banana")
trie.insert("orange")

print(trie.search("apple"))  # Output: True
print(trie.search("banana"))  # Output: True
print(trie.search("orange"))  # Output: True
print(trie.search("grape"))  # Output: False

print(trie.starts_with("app"))  # Output: True
print(trie.starts_with("ban"))  # Output: True
print(trie.starts_with("ora"))  # Output: True
print(trie.starts_with("gr"))  # Output: False

trie.delete("banana")
print(trie.search("banana"))  # Output: False
````

## 27. Coin Change `Medium`
- You are given an integer array coins representing coins of different denominations and an integer amount representing a total amount of money.
- Return the fewest number of coins that you need to make up that amount. If that amount of money cannot be made up by any combination of the coins, return -1

Test Case 1:
```python
coins = [1, 2, 5]
amount = 11
# Expected output: 3
```

Test Case 2:
```python
coins = [2]
amount = 3
# Expected output: -1
```

Test Case 3:
```python
coins = [1, 3, 4, 5]
amount = 7
# Expected output: 2
```

Test Case 4:
```python
coins = [2, 5, 10]
amount = 16
# Expected output: 4
```

Test Case 5:
```python
coins = [1, 2, 5]
amount = 18
# Expected output: 5
```


````python
def coinChange(coins, amount):
    if amount == 0:
        return 0
    
    if amount < 0:
        return -1
    
    min_coins = float('inf')
    
    for coin in coins:
        subproblem = coinChange(coins, amount - coin)
        
        if subproblem >= 0 and subproblem < min_coins:
            min_coins = subproblem + 1
    
    if min_coins == float('inf'):
        return -1
    
    return min_coins
````

````python
coins = [1, 2, 5]
amount = 11
result = coinChange(coins, amount)
print(result)  # Output: 3


coins = [2]
amount = 3
result = coinChange(coins, amount)
print(result)  # Output: -1

coins = [1, 3, 4, 5]
amount = 7
result = coinChange(coins, amount)
print(result)  # Output: 2

coins = [2, 5, 10]
amount = 16
result = coinChange(coins, amount)
print(result)  # Output: 4

coins = [1, 2, 5]
amount = 18
result = coinChange(coins, amount)
print(result)  # Output: 5
````

- You must write an algorithm that runs in `O(n)` time and without using the division operation.

````python
def product_array(nums):

    left_product = [1] * len(nums)
    right_product = [1] * len(nums)
    
    product = 1
    for i in range(1,len(nums)):
        left_product[i] = left_product[i-1] * nums[i-1]
    
    product = 1
    for i in range(len(nums)-1, 0, -1):
        right_product[i-1] = right_product[i] * nums[i]
        
    answer = []
    for i,j in zip(right_product, left_product):
        answer.append(i * j)

    return answer
    
nums = [1,2,3,4,5]
print(product_array(nums))
````

## 29. Min Stack `Medium`
- Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.
- Implement the MinStack class:
- MinStack() initializes the stack object.
- void push(int val) pushes the element val onto the stack. void pop() removes the element on the top of the stack.
- int top() gets the top element of the stack.
- int getMin() retrieves the minimum element in the stack.
- You must implement a solution with O(1) time complexity for each function.:

````python
class MinStack():
    
    def __init__ (self):
        self.data = []
        
    def push(self, value):
        queue = []
        while not self.empty() and self.top() > value:
            queue.append(self.pop())
        
        self.data.append(value)
        while len(queue) > 0:
            self.data.append(queue.pop())
            
    def empty(self):
        return len(self.data) == 0
    
    def pop(self):
        if self.empty():
            raise Exception('Empty Stack')
        val = self.data[-1]
        del self.data[-1]
        return val
    
    def top(self):
        if self.empty():
            raise Exception('Empty Stack')
        return self.data[-1]
    
    def get_min(self):
        if self.empty():
            raise Exception('Empty Stack')
        return self.data[0]    

    def __str__(self):
        return str(self.data)
    
````

````python
stack = MinStack()
print(stack.empty())
stack.push(10)
stack.push(7)
stack.push(1)
stack.push(4)
print(stack)
print(stack.top())
print(stack.get_min())
print(stack.pop())
print(stack)
print(stack.top())
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack, stack.empty())
# print(stack.pop()) # Empty Stack
````

## 30. Validate Binary Search Tree `Medium`
- Given the root of a binary tree, determine if it is a valid binary search tree (BST).


1. For any node in the tree, all nodes in its left subtree have values less than its value.
2. For any node in the tree, all nodes in its right subtree have values greater than its value.
3. Both the left and right subtrees of every node are also valid BSTs.


````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursive(self.root, data)

    def _insert_recursive(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursive(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursive(node.right, data)
        else:
            # Ignore duplicate values (assuming no duplicates are allowed)
            pass
    
def is_valid(root):
    if root == None:
        return True
    
    if root.left is not None and root.left.data >= root.data:
        return False

    if root.right is not None and root.right.data < root.data:
        return False
    
    return is_valid(root.left) and is_valid(root.right)    
````
